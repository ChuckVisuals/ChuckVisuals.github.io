```python
import pandas as pd
import numpy as np
df = pd.read_csv("animes.csv")
import matplotlib.pyplot as plt
import scipy
import re
# Remove duplicates
df = df.drop_duplicates()
# changing columns types from string to int
df['members'] = df['members'].astype(int)
df['popularity'] = df['popularity'].astype(int)
# changing string back into string array with each genre being its own element in array
df['genre'] = df['genre'].apply(eval)
#dropping the image url and link from table since it won't be needed
df = df.drop('img_url', axis=1)
df = df.drop('link', axis=1)
# converting string into two dates and only getting the first date
df['aired'] = df['aired'].str.split(' to ').str[0]
#function that will format the string into a datetime object to be comparable, also adds in missing data if missing month and or day.
def parse_date(date_str):
    formats = [r'^\w+ \d+\, \d{4}$', r'^\w+ \d{4}$', r'^\d{4}$']
    for fmt in formats:
      x = re.search(fmt, date_str)
      if (x):
        x = str(x.group())
        if fmt == r'^\w+ \d{4}$':
          x = x.replace(" ", " 1, ")
        elif fmt == r'^\d{4}$':
          x = "January 1, "+ x
      return pd.to_datetime(x)
    return pd.NaT
df['aired'] = df['aired'].apply(parse_date).dt.year
df['synopsis'] = df['synopsis'].fillna(' ')
df = df.dropna()
display(df)
df.to_csv('d2.csv')
# df_genres = df.explode('genre')
# genre_counts = df_genres['genre'].value_counts()
# df = df.dropna()
# display(genre_counts)
# display(len(df_genres['genre']))
# unique_genres = df_genres['genre'].unique()

# # Create new genre columns with 0s
# for genre in unique_genres:
#   df[genre] = 0

# # Fill 1s for matching genres in each row
# for index, row in df_genres.iterrows():
#   df.loc[index, row['genre']] = 1

# # Drop the original 'genre' column (optional)
# print(df)
# df.dropna()
# df = df.loc[:, (df.columns.astype(str) != 'nan') & (df.columns.notna()) & (df.columns.str.isalnum())]
# df = df.dropna()
```

```python
def remove_non_str_nan(df):
  """
  Removes columns from a DataFrame where the column title is not a string or is NaN.

  Args:
      df (pandas.DataFrame): The DataFrame to filter.

  Returns:
      pandas.DataFrame: The DataFrame with filtered columns.
  """
  return df.loc[:, (df.columns.astype(str) != 'nan') & (df.columns.notna()) & (df.columns.str.isalnum())]

filtered_df = remove_non_str_nan(df.copy())
print(filtered_df)
```
